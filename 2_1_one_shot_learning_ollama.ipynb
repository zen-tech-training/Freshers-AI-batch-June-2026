```python
import ollama
```

```python
one_shot_prompt = (
            f"Example:\n"
            f"Q: Translate 'Hello, how are you?' into Hindi.\n"
            f"A: Aap kaise ho?\n\n"
            f"Now answer this:\n"
            f"Q: Translate 'What is your name?' into Hindi.\nA:"
        )
response = ollama.chat(
    model="gpt-oss:120b-cloud",
    messages=[
        {"role": "user", "content": one_shot_prompt}
    ]
)
print(response['message']['content'])
```
