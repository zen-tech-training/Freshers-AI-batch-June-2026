```python
# Example 1 - Basic Embedding
import ollama
from dotenv import find_dotenv, load_dotenv
import os
from langchain_ollama import OllamaEmbeddings
from langchain_community.vectorstores import FAISS

def get_embedding(text, model="nomic-embed-text"):
    text = text.replace("\n", " ")
    response = ollama.embeddings(prompt=text, model=model)
    print(response)
    return response.embedding

# Example usage:
input_text = "Hello, world! This is a test sentence for embedding."
embedding = get_embedding(input_text)

print(f"Input Text: {input_text}")
print(f"Embedding (first 5 values): {embedding[:5]}...")
print(f"Embedding dimension: {len(embedding)}")
```

```python
embedding_model = OllamaEmbeddings(model="nomic-embed-text")

input_text = "Hello, world! This is a test sentence for embedding."
embedding = embedding_model.embed_query(input_text)

print(f"Input Text: {input_text}")
print(f"Embedding (first 5 values): {embedding[:]}...")
print(f"Embedding dimension: {len(embedding)}")
```

```python
input_text = "I like music"
words = input_text.split()
for word in words:
    embedding = embedding_model.embed_query(input_text)
    print(f"Input Text: {input_text}")
    print(f"Embedding (first 5 values): {embedding[:]}...")
    print(f"Embedding dimension: {len(embedding)}")

embedding = embedding_model.embed_query(input_text)

print(f"Input Text: {input_text}")
print(f"Embedding (first 5 values): {embedding[:]}...")
print(f"Embedding dimension: {len(embedding)}")
```

```python
vector_store = {}
input_text = "I like music"
words = input_text.split()
for word in words:
    embedding = embedding_model.embed_query(word)
    vector_store[word] = embedding
    print(f"Input Text: {word}")
    print(f"Embedding (first 5 values): {embedding[:]}...")
    print(f"Embedding dimension: {len(embedding)}")

embedding = embedding_model.embed_query(input_text)
print(vector_store)
```

```python

texts = ["I like music", "I love cooking", "My name is Anand"]
vector_store = FAISS.from_texts(texts, embedding_model)
print("Vector store created successfully")
print(vector_store)
```
