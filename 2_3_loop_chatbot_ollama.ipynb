```python
import ollama
```

```python
while True:
    user_prompt = input('Enter the prompt (type quit to exit)')
    if user_prompt.lower() == 'quit':
        break

    response = ollama.chat(
        model="gpt-oss:120b-cloud",
        messages=[
            {"role": "user", "content": user_prompt}
        ]
    )
    print(response['message']['content'])
```
