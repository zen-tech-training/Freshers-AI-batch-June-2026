```python
import ollama
```

```python
few_shot_prompt = """
    Examples:
    Message: My laptop hangs frequently
    Category: Laptop

    Message: Food quality is poor in office canteen
    Category: Food

    Now classify the following message:
    Message: Bus driver is a rash driver
    Category: 
"""
response = ollama.chat(
    model="gpt-oss:120b-cloud",
    messages=[
        {"role": "user", "content": few_shot_prompt}
    ]
)
print(response['message']['content'])
```
