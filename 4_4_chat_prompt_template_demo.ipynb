```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
from dotenv import find_dotenv, load_dotenv
import os

env_path = find_dotenv()
if not env_path:
    raise FileNotFoundError(".env file not found.")

load_dotenv(env_path)
apiKey = os.getenv("OLLAMA_API_KEY")
os.environ['OLLAMA_API_KEY'] = apiKey
```

```python
model = ChatOllama(model="gpt-oss:120b-cloud")
# Loop to ask multiple questions
chat_history = []
while True:
    question = input("\nEnter the question (or type 'exit' to quit): ")
    if question.lower() in ['exit', 'quit']:
        print("Exiting... Have a great day!")
        break
    chat_history.append(SystemMessage("You are a helpful AI assistant"))
    chat_history.append(HumanMessage(question))
    response = model.invoke(question)
    chat_history.append(AIMessage(response.content))

    # Show the answer
    print("\nAnswer:", response.content)
    print(response.content)

print(chat_history)
```

```python
# country will not be replaced with the value
chat_template = ChatPromptTemplate([
    SystemMessage("You are a helpful AI assistant"),
    HumanMessage("Which is the capital city of {country}?")
])

# country will be replaced with the value
chat_template = ChatPromptTemplate([
    ("system", "You are a helpful AI assistant"),
    ("user", "Which is the capital city of {country}?")
])


prompt = chat_template.invoke({"country": "Sweden"})
print(prompt)

response = model.invoke(prompt)
print(response.content)
```

```python
prompt = ChatPromptTemplate.from_messages(
    [
 ("system", "You are a helpful AI assistant"),
 MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{user_input}")    
])

chat_history = ([
    HumanMessage(content="What is capital city of India?"),
    AIMessage(content="New Delhi is a capital city of India")
])

messages = prompt.format_messages(chat_history=chat_history, user_input="Which is the famous landmark there?")
response = model.invoke(messages)
print(response.content)
```
