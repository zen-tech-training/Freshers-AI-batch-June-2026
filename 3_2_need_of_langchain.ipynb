```python
import ollama
from transformers import pipeline
from openai import OpenAI
import os
from dotenv import load_dotenv, find_dotenv
from huggingface_hub import InferenceClient

env_path = find_dotenv()
if not env_path:
    raise FileNotFoundError(".env file not found.")

load_dotenv(env_path)
apiKey = os.getenv("OPENAI_API_KEY")
hfApiKey = os.getenv("HF_API_KEY")
```

```python
response = ollama.chat(
    model="gpt-oss:120b-cloud",
    messages=[
        {"role": "user", "content": "Which is the capital city of India?"}
    ]
)
print(type(response))
print(response['message']['content'])
```

```python
client = OpenAI(api_key=apiKey)
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": "Which is the capital city of India?"}
    ]
)
print(type(response))
print(response.choices[0].message.content)
```

```python
sentiment_analyzer = pipeline('sentiment-analysis', model='distilbert-base-uncased-finetuned-sst2')
prompt = 'Explain the need of Langchain'
response = sentiment_analyzer(prompt)
print(response)
print(type(response))
```

```python
hf_client = InferenceClient(token=hfApiKey)
response = hf_client.chat.completions.create(
    model="distilbert-base-uncased-finetuned-sst2",
    messages=[
        {"role": "user", "content": "Which is the capital city of India?"}
    ]
)
print(type(response))
print(response)
```
