```python
import tiktoken
encoding = tiktoken.get_encoding('cl100k_base')
#text = 'I work in IT industry. I like mechanical industry'
text = 'Generative AI (Gen AI) is a branch of artificial intelligence focused on content creation'
tokens = encoding.encode(text)
print(tokens)
print(len(tokens))
print('Decoded text ', encoding.decode(tokens))

for token in tokens:
    print(f"Token ID: {token}, Input test: {encoding.decode([token])}")
```
