```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
from dotenv import find_dotenv, load_dotenv
import os

env_path = find_dotenv()
if not env_path:
    raise FileNotFoundError(".env file not found.")

load_dotenv(env_path)
apiKey = os.getenv("OLLAMA_API_KEY")
os.environ['OLLAMA_API_KEY'] = apiKey
```

```python
# Example 1 - Single input variable

model = ChatOllama(model="gpt-oss:120b-cloud")
# Loop to ask multiple questions
while True:
    topic = input("\nEnter the concept (or type 'exit' to quit): ")

    if topic.lower() in ['exit', 'quit']:
        print("Exiting... Have a great day!")
        break

    template = PromptTemplate(
        input_variables=["topic"],
        template="Explain the concept of {topic} in simple terms."
    )

    # Fill the template with a value
    filled_prompt = template.format(topic=topic)
    print(filled_prompt)

    chat_response = model.invoke(filled_prompt)

    # Show the answer
    print("\nAnswer:", chat_response.content)    
    print(chat_response.content)
```

```python
# Example 2 - Multiple input variables

model = ChatOllama(model="gpt-oss:120b-cloud")
template = """You are a helpful assistant.
Summarize the following article in {word_count} words:
{text}"""

template = PromptTemplate(
    input_variables=["word_count", "text"],
    template=template
)

text = """
Generative AI (GenAI) is a type of artificial intelligence that creates new content—text, images, audio, video, or code—by learning patterns from existing data.
What is GenAI?
Generative AI, or GenAI, refers to AI systems that generate original content rather than merely analyzing or classifying data. Unlike traditional AI, which predicts or categorizes information, GenAI models learn the underlying patterns in large datasets and produce outputs that resemble real-world examples but are not exact copies. For instance, a GenAI model trained on thousands of paintings can create a new artwork in the style of Van Gogh that never existed before.
"""
# Fill the template with a value
filled_prompt = template.format(word_count=5, text=text)
print(filled_prompt)

chat_response = model.invoke(filled_prompt)

# Show the answer
print("\nAnswer:", chat_response.content)    
print(chat_response.content)
```

```python
# Example 3 - Mustache input variables
# Refer https://docs.langchain.com/langsmith/prompt-template-format

model = ChatOllama(model="gpt-oss:120b-cloud")

input = {
    "name": 'Sachin Tendulkar',
    "team": "Indian cricket team"
}
template = "Please find information about {name} working with {team}"

template = PromptTemplate(
    input_variables=["name", "team"],
    template=template
)

# Fill the template with a value
filled_prompt = template.format(**input)
print(filled_prompt)

chat_response = model.invoke(filled_prompt)

# Show the answer
print("\nAnswer:", chat_response.content)    
print(chat_response.content)
```
